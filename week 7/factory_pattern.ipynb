```python
from abc import ABC, abstractmethod

# Abstract Factory
class Factory(ABC):
    @abstractmethod
    def create_product(self, kind=None):
        pass

# Concrete Factory
class AnimalFactory(Factory):
    def create_product(self, kind=None):
        if kind == "dog":
            return Dog()
        elif kind == "cat":
            return Cat()
        else:
            print("Unknown animal type!")
            return None

# Abstract Product
class Animals(ABC):
    @abstractmethod
    def run(self):
        pass

# Concrete Products
class Dog(Animals):
    def run(self):
        print("I'm a Dog, I can run!!")

class Cat(Animals):
    def run(self):
        print("I'm a Cat, I can run!!")

# âœ… Correct Client Code
factory = AnimalFactory()
dog = factory.create_product("dog")
if dog: 
    dog.run()
```
